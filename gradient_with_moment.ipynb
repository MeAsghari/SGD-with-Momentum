```python
def funclog(x):
  return torch.log(1 + torch.abs(x))**(2 + torch.sin(x))
```

```python
def gradient_descent_with_momentum(func, xi, eta, N , beta):
  xis = []
  yis = []
  mt0 = 0

  for iter in range(N):
    func(xi).backward()

    mt1 = beta*mt0 + (1-beta)*xi.grad
    xi.data -= eta * mt1
    xi.grad.zero_()

    xi_copy = xi.clone()
    xis.append(xi_copy.data)
    yis.append(func(xi_copy.data))

    mt0 = mt1

  return xis , yis
```

```python
betas = [0.96 , 0.9 , 0.7 , 0.5 , 0.4 , 0.1]
X = torch.linspace(-10.5, 10.5, 100)
fig , axe = plt.subplots(2 , 3 , figsize=(15 , 9))
axes = axe.flatten()

for i , beta in enumerate(betas):
  x = torch.tensor(7., requires_grad=True)
  eta = 0.15
  N = 100

  xs , ys = gradient_descent_with_momentum(funclog, x, eta, N , beta=beta)
  xs.insert(0 , 7.)
  ys.insert(0 , funclog(torch.tensor(7.)))

  axes[i].plot(X , funclog(X) , color='b')
  axes[i].plot(xs , ys , color='r',  marker="o")
  axes[i].plot(xs[-1] , ys[-1] , color='g' , marker='*' , ms=15)
  axes[i].set_title(f'lr={eta} / Num Epoch={N} / beta={beta}' , fontfamily='Serif')
```
